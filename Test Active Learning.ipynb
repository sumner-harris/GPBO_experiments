```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import os

%load_ext autoreload
%autoreload 1
%aimport active_learning
from active_learning import step_GP, append_data, save_diagnostics, predict_posterior, plot_2D
from testing_functions import f2
torch.manual_seed(10)

# Parameters
MAXIMIZE = True
ACQ = 'EI'
UCB_BETA = 8 #only used if ACQ is UCB
BATCH = 1
N_STEPS = 20
SAVE_PATH = 'test_results'
FILE_NAME = 'testname'
DISCRETE = False

# Domain and test grid
bounds = torch.tensor([[-4.0, -4.0], [4.0, 4.0]], dtype=torch.float64)
grid_lin = torch.linspace(-4, 4, 20, dtype=torch.float64)    # 40 points
X0, X1  = torch.meshgrid(grid_lin, grid_lin, indexing="ij") # each (40,40)
X_test  = torch.stack([X0.ravel(), X1.ravel()], dim=-1)      # (1600, 2)

data  = torch.load("starting_data.pt")
X     = data['X']
Y     = data['Y']
Y_var = data['Y_var']
mean  = data['mean']
var   = data['var']
acq   = data['acq']

for i in range(1, N_STEPS+1):
    print(i)
    # 4.1 Propose next point (batch=1), here we MINIMIZE the function:
    x_next, mean, var, acq = step_GP(
        X=X, Y=Y, X_test=X_test, bounds=bounds,
        Y_var=Y_var,
        acq_func=ACQ, beta=UCB_BETA,
        batch=BATCH,
        maximize=MAXIMIZE,
        discrete=DISCRETE,
        num_restarts=10,
        raw_samples=100,
    )

    # 4.2 Run real experiment:
    torch.save({'x_next':x_next},os.path.join(SAVE_PATH,f"suggested_experiment_conditions_{i}.pt"))
    y_next = f2(x_next, noise_level=5)
    Y_var_new = torch.ones_like(y_next)*5

    # 4.3 Append & save
    X, Y = append_data(
        X, Y, Y_var,
        X_new=x_next,
        Y_new=y_next,
        Y_var_new=Y_var_new,
        mean=mean,
        var=var,
        acq=acq,
        save_path=SAVE_PATH,
        #base_filename=FILE_NAME+f'_step_{i}',
        base_filename=FILE_NAME,
        step=i
    )
    
    Y_var = torch.ones_like(Y)*100
    
    save_diagnostics(
        X_test,X,Y,
        mean,acq,
        maximize=MAXIMIZE,
        save_path=SAVE_PATH,         
        base_filename=FILE_NAME,
        step=i
        )

    print(f"Step {i:2d} â†’ x={x_next},  y={y_next}")
    plot_2D(X,Y,X0,X1,mean,var,acq)
```

```python
from natsort import natsorted

diagnostic_files = natsorted([
    os.path.join(SAVE_PATH, fname)
    for fname in os.listdir(SAVE_PATH)
    if "_diagnostics" in fname
])

# 2) empty lists to collect each quantity
X_best_list       = []
Y_best_list       = []
X_best_pred_list  = []
Y_best_pred_list  = []
acq_max_list      = []

# 3) load each file and append
for fn in diagnostic_files:
    data = torch.load(fn)
    X_best_list.append(data['X_best'])
    Y_best_list.append(torch.tensor(data['Y_best']))
    X_best_pred_list.append(data['X_best_pred'])
    Y_best_pred_list.append(torch.tensor(data['Y_best_pred']))
    acq_max_list.append(data['acq_max'])

# 4) stack into tensors
X_best       = torch.stack(X_best_list)        # shape (M, 2)
Y_best       = torch.stack(Y_best_list).unsqueeze(-1)       # (M,1)
X_best_pred  = torch.stack(X_best_pred_list)   # (M, 2)
Y_best_pred  = torch.stack(Y_best_pred_list).unsqueeze(-1)  # (M,1)
acq_max      = torch.stack(acq_max_list).unsqueeze(-1)      # (M,1)

# 5) simple plot: true vs predicted best over file index
idx = torch.arange(len(diagnostic_files))
plt.figure(figsize=(6,4))
plt.plot(idx, Y_best.squeeze(-1).numpy(),       label='Y_best (true)',   marker='o')
plt.plot(idx, Y_best_pred.squeeze(-1).numpy(),  label='Y_best_pred (GP)', marker='x')
plt.xlabel("Step")
plt.ylabel("Best objective")
plt.legend()
plt.tight_layout()
plt.show()

plt.plot(idx, acq_max.exp(),'o-')
```
